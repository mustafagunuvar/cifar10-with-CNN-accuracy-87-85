## Data Preprocessing

```python
X_train[5].min()
```

```python
X_train[5].max()
```

```python
# Normalize image pixels
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
X_train[5].min()
```

```python
X_train[5].max()
```

```python
X_train.shape, X_test.shape
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
Y_train = to_categorical(y_train, 10)
Y_test = to_categorical(y_test, 10)
Y_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, Flatten, Dropout
```

## Model Training

```python
# Model - 1 

model = Sequential()

model.add(Conv2D(filters=32, kernel_size=(3,3),input_shape=(32, 32, 3), activation='relu', padding='same'))

model.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu', padding='same'))

model.add(MaxPool2D(pool_size=(2, 2)))

model.add(Conv2D(filters=32, kernel_size=(3,3), activation='relu', padding='same'))

model.add(MaxPool2D(pool_size=(2, 2)))

model.add(Flatten())

model.add(Dense(128, activation='relu'))

model.add(Dense(64, activation='relu'))

model.add(Dense(10, activation='softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
from tensorflow.keras.optimizers import AdamW
```

```python
# Model - 2

model_2 = Sequential()

model_2.add(Conv2D(filters=64, kernel_size=(3,3),input_shape=(32, 32, 3), activation='relu', padding='valid'))

model_2.add(Conv2D(filters=32, kernel_size=(3,3), activation='relu', padding='valid'))

model_2.add(MaxPool2D(pool_size=(2, 2)))

model_2.add(Conv2D(filters=32, kernel_size=(3,3), activation='relu', padding='valid'))

model_2.add(MaxPool2D(pool_size=(2, 2)))

model_2.add(Flatten())

model_2.add(Dense(128, activation='relu'))

model_2.add(Dense(64, activation='relu'))

model_2.add(Dense(128, activation= "relu"))

model_2.add(Dense(10, activation='softmax'))

optimizer = AdamW(learning_rate= 0.003)

model_2.compile(loss='categorical_crossentropy',
              optimizer= optimizer,
              metrics=['accuracy'])
```

```python
from tensorflow.keras.layers import BatchNormalization
```

```python
# Model - 3

model_3 = Sequential()

model_3.add(Conv2D(filters=128, kernel_size=(3,3),input_shape=(32, 32, 3), activation='relu', padding='valid'))

model_3.add(BatchNormalization())

model_3.add(Conv2D(filters=128, kernel_size=(3,3), activation='relu', padding='valid'))

model_3.add(BatchNormalization())

model_3.add(MaxPool2D(pool_size=(2, 2)))

model_3.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu', padding='valid'))

model_3.add(BatchNormalization())

model_3.add(MaxPool2D(pool_size=(2, 2)))

model_3.add(Flatten())

model_3.add(Dense(128, activation='relu'))

model_3.add(Dropout(0.5))

model_3.add(Dense(64, activation='relu'))

model_3.add(Dropout(0.5))

model_3.add(Dense(128, activation= "relu"))

model_3.add(Dense(10, activation='softmax'))

optimizer = AdamW(learning_rate= 0.0001)

model_3.compile(loss='categorical_crossentropy',
              optimizer= optimizer,
              metrics=['accuracy'])
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import AdamW
from tensorflow.keras.callbacks import EarlyStopping

model4 = Sequential()

model4.add(Conv2D(32, (3,3), padding='same', activation='relu', input_shape=(32,32,3)))
model4.add(BatchNormalization())
model4.add(Conv2D(32, (3,3), padding='same', activation='relu'))
model4.add(BatchNormalization())
model4.add(MaxPool2D(pool_size=(2,2)))
model4.add(Dropout(0.3))

model4.add(Conv2D(64, (3,3), padding='same', activation='relu'))
model4.add(BatchNormalization())
model4.add(Conv2D(64, (3,3), padding='same', activation='relu'))
model4.add(BatchNormalization())
model4.add(MaxPool2D(pool_size=(2,2)))
model4.add(Dropout(0.5))

model4.add(Conv2D(128, (3,3), padding='same', activation='relu'))
model4.add(BatchNormalization())
model4.add(Conv2D(128, (3,3), padding='same', activation='relu'))
model4.add(BatchNormalization())
model4.add(MaxPool2D(pool_size=(2,2)))

model4.add(Conv2D(256, (3,3), padding='same', activation='relu'))
model4.add(BatchNormalization())
model4.add(Conv2D(256, (3,3), padding='same', activation='relu'))
model4.add(BatchNormalization())
model4.add(MaxPool2D(pool_size=(2,2)))

model4.add(Flatten())

model4.add(Dense(256, activation='relu'))
model4.add(BatchNormalization())

model4.add(Dense(128, activation='relu'))
model4.add(BatchNormalization())
model4.add(Dropout(0.5))

model4.add(Dense(10, activation='softmax'))

optimizer = AdamW(learning_rate=0.0007)
model4.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])

early_stop = EarlyStopping(monitor='val_loss', patience=7)

datagen = ImageDataGenerator(
    rotation_range=15,      
    width_shift_range=0.1,  
    height_shift_range=0.1, 
    horizontal_flip=True,  
    zoom_range=0.2          
)

datagen.fit(X_train)

history = model4.fit(
    datagen.flow(X_train, Y_train, batch_size=32),
    validation_data=(X_test, Y_test),
    epochs=100,
    callbacks=[early_stop],
    verbose=1
)
```

```python
plt.figure(figsize=(15, 6))
plt.plot(summary.accuracy, label="accuracy")
plt.plot(summary.val_accuracy, label="val_accuracy")
plt.legend(loc="upper left")
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.show()
```
